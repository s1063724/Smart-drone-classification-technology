```python
# -*- coding: utf-8 -*-
import folium# 匯入 folium 套件

# 建立地圖與設定位置
fmap = folium.Map(location=[24.2258027,120.5750026], zoom_start=16)
fmap  # 在notebook中顯示地圖
center_pos = [24.227254, 120.581077]
lat, lon = center_pos
fmap = folium.Map(location=center_pos, zoom_start=17)

#定義圓形方法
def add_circle(fmap, pos, text, radius: int, color='red'):
    x, y = pos
    folium.Circle(pos, radius, color=color, fill=True, 
                  fill_opacity=0.7 # 設定透明度
                 ).add_to(fmap)#color(parameter)=color(variable)->color=red的意思
```

```python
import csv
#開啟csv檔
with open('map1.csv', 'r', newline='', encoding='utf-8') as fp:
    data = [dict(row) for row in csv.DictReader(fp)]#DictReader可將檔案轉為 Python 的 dictionary 格式
```

```python
#python dictionary
influences = {
    '0': {'radius': 5, 'color':'green'},
    '1': {'radius': 10, 'color':'yellow'},
    '2': {'radius': 15, 'color':'red'}
}
influences
```

```python
for item in data:
    position = [item['lat'], item['lon']]
    title = item['title']
    add_circle(fmap, position, title, **influences[item['influence']])# 雙星號有此作用->spread operator
```
